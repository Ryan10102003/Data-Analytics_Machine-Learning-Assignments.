# Feature Engineering Assignment


1. What is a parameter?
 - A parameter is a variable used by a model to make predictions, learned from the training data (e.g., weights in linear regression).


2. What is correlation?
What does negative correlation mean?
 - Correlation measures the strength and direction of a relationship between two variables.
Negative correlation means that as one variable increases, the other decreases.


3. Define Machine Learning. What are the main components in Machine Learning?
 - Machine learning is a field of AI where systems learn from data to make predictions or decisions without being explicitly programmed.
The main components are data, model, algorithm, and objective function.


4. How does loss value help in determining whether the model is good or not?
 - A lower loss value indicates the model’s predictions are closer to the actual values, meaning the model is performing well.
A high loss means poor prediction accuracy and the model needs improvement.


5. What are continuous and categorical variables?
 - Continuous variables can take any numerical value within a range (e.g., height, weight).
Categorical variables represent distinct groups or categories (e.g., gender, color).


6. How do we handle categorical variables in Machine Learning? What are the common techniques?
 - We handle categorical variables by converting them into numerical form.
Common techniques include label encoding and one-hot encoding.


7. What do you mean by training and testing a dataset?
 - Training a dataset means using data to teach the model patterns.
Testing a dataset means evaluating the model’s performance on unseen data.


8. What is sklearn.preprocessing?
 - `sklearn.preprocessing` is a module in Scikit-learn used for scaling, transforming, and encoding data before feeding it into a model.


9. What is a Test set?
 - A test set is a portion of data used to evaluate the model's performance on unseen data after training.


10. How do we split data for model fitting (training and testing) in Python?
How do you approach a Machine Learning problem?
 - We split data using functions like `train_test_split` from Scikit-learn.
Approach: understand problem, collect data, preprocess, choose model, train, evaluate, and tune.


11. Why do we have to perform EDA before fitting a model to the data?
 - EDA helps understand data patterns, detect outliers, and find relationships, improving model accuracy.


12. What is correlation?
 - Correlation shows how strongly two variables are related and their direction (positive or negative).


13. What does negative correlation mean?
 - Negative correlation means when one variable increases, the other decreases.


14. How can you find correlation between variables in Python?
 - Use `df.corr()` in pandas to find correlation between variables in Python.


15. What is causation? Explain difference between correlation and causation with an example?
 - Causation means one event directly causes another.
Correlation shows a relationship but doesn’t prove cause; e.g., ice cream sales and drowning incidents correlate but ice cream doesn’t cause drowning.


16. What is an Optimizer? What are different types of optimizers? Explain each with an example?
 - An optimizer adjusts model parameters to minimize loss during training.
Types include Gradient Descent (updates parameters using gradients), Adam (combines momentum and adaptive learning rates), and RMSprop (uses moving average of squared gradients).


17. What is sklearn.linear_model?
 - `sklearn.linear_model` is a Scikit-learn module providing linear models like Linear Regression and Logistic Regression for prediction tasks.


18. What does model.fit() do? What arguments must be given?
 - `model.fit()` trains the model using input features `X` and target labels `y`.
Arguments: `X` (training data), `y` (target values).


19. What does model.predict() do? What arguments must be given?
 - `model.predict()` makes predictions using the trained model on new data.
Argument: `X` (input features for prediction).


20. What are continuous and categorical variables?
 - Continuous variables are numeric and can take any value within a range (e.g., temperature).
Categorical variables represent distinct groups or labels (e.g., color, gender).


21. What is feature scaling? How does it help in Machine Learning?
 - Feature scaling standardizes data values to a similar range, improving model training speed and accuracy.
It prevents features with large ranges from dominating the model.


22. How do we perform scaling in Python?
 - We perform scaling using Scikit-learn’s `StandardScaler` or `MinMaxScaler` on the data.
Example: `scaler = StandardScaler(); X_scaled = scaler.fit_transform(X)`


23. What is sklearn.preprocessing?
 - `sklearn.preprocessing` is a Scikit-learn module for transforming and scaling data before modeling.


24. How do we split data for model fitting (training and testing) in Python?
 - Use `train_test_split` from Scikit-learn to split data into training and testing sets.
Example: `X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)`


25. Explain data encoding?
 - Data encoding converts categorical data into numerical format for machine learning models.
Examples include label encoding and one-hot encoding.
